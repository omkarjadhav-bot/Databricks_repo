# Create temporary view
1. Exist only within current notebook or spark session
2. Other notebook can not access it
3. Disappers when detach your notebook from cluster, session ends,   cluster restart

```sql
CREATE OR REPLACE TEMPORARY VIEW tv_customers AS
SELECT
  *,
  input_file_name() AS file_path
FROM json.`/Volumes/gizmobox/landing/operational_data/customers/`
```

```sql
select * from tv_customers;
```

# Create Global temporary view
1. Accessibal across all notebooks and spark session on the same cluster
2. Lives under special database: `global_temp`
3. Disappears when cluster stops or restarts


```sql
CREATE OR REPLACE GLOBAL TEMPORARY VIEW gtv_customers AS
SELECT
  *,
  input_file_name() AS file_path
FROM json.`/Volumes/gizmobox/landing/operational_data/customers/`
```

```sql
select * from global_temp.gtv_customers;
```
